**Convert responses to numeric**

```python
#Remove timestamp and comments columns
responses_num = responses.iloc[:,1:25]
responses_num.head()
```

```python
rosurvey_dict1 = {
    'NOT IMPORTANT at All' : 1,
    'Of LITTLE IMPORTANCE' : 2,
    'Of AVERAGE IMPORTANCE' : 3,
    'VERY IMPORTANT' : 4,
    'ABSOLUTELY ESSENTIAL' : 5
}
```

```python
rosurvey_dict2 = {
    'NEVER' : 1,
    'LESS THAN ONCE PER WEEK on average' : 2,
    'ONCE PER WEEK on average' : 3,
    '2 or 3 TIMES PER WEEK on average' : 4,
    '4 OR MORE TIMES PER WEEK on average' : 5
}
```

```python
rosurvey_dict3 = {
    'I am DEFINITELY NOT READY to attend any in-person events for the foreseeable future.' : 1,
    'I am PROBABLY NOT READY to attend in-person events.' : 2,
    'I am SOMEWHAT READY to attend in-peson events' : 3,
    'I am PROBABLY READY to attend in-person events.' : 4,
    'I am DEFINITELY READY to attend in-person events.' : 5
}
```

```python
#Map numerics to responses for rating columns for analysis

responses_num = responses_num.replace(rosurvey_dict1)
responses_num.head()
```

```python
#Map numerics to responses in 'avg_weekly_outings' column for analysis

responses_num.avg_weekly_outings = responses_num.avg_weekly_outings.replace(rosurvey_dict2)
responses_num.avg_weekly_outings.head()
```

```python
#Map numerics to responses in 'return_readiness' column for analysis

responses_num.return_readiness = responses_num.return_readiness.replace(rosurvey_dict3)
responses_num.return_readiness.head()
```
