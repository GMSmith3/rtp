```python
#Create list of email addresses for those who have already responded
respondents = clicks_report_Aug26.email.to_list()
type(respondents)
```

```python
#Create list of email addresses for those in survey recipient list
recipient_emails = survey_list['Contact: Email'].to_list()
type(recipient_emails)
```

```python
#Create a list of emails for those in the survey list who have not yet responded
follow_ups = [e for e in recipient_emails if e not in respondents]
len(follow_ups)
```

```python
#Create df containing only the survey_list members who have not yet responded
survey_list_follow_ups = survey_list[survey_list['Contact: Email'].isin(follow_ups)].reset_index(drop=True)
survey_list_follow_ups.head()
```

```python
#Export follow up email list for launch
survey_list_follow_ups.to_csv('Mobile Bartender Survey Follow Up Email List.csv')
```
